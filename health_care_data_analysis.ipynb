```python
import pandas as pd
```

```python
data=pd.read_csv('/content/healthcare-dataset-stroke-data.csv')
```

```python
data
```

```python
data=data.drop(['work_type','Residence_type'],axis=1)
data
```

Finding Null Values

```python
data.isna().sum()
```

droping null values

```python
data=data.drop(['bmi'],axis=1)
```

```python
data
```

Getting data type of data

```python
data.info()
```

Converting objects into integers

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data['gender']=le.fit_transform(data['gender'])
data['ever_married']=le.fit_transform(data['ever_married'])
data['smoking_status']=le.fit_transform(data['smoking_status'])
```

```python
data
```

```python
ind=data.iloc[:,0:-1]
```

```python
ind
```

```python
dep=data.iloc[:,-1]
```

```python
dep
```

Training dependent and independent data

```python
from sklearn.model_selection import train_test_split
```

```python
ind_train,ind_test,dep_train,dep_test=train_test_split(ind,dep,test_size=0.2)
```

```python
ind_train
```

```python
ind_test
```

```python
dep_train
```

```python
dep_test
```

**Choosing algoritham**

Using Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(ind_train,dep_train)
```

```python
dep_pred=dtc.predict(ind_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(dep_test,dep_pred)
```

```python
cm
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
sns.heatmap(cm,
            annot=True,
            xticklabels=['pass','fail'],
            yticklabels=['pass','fail'])
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

```python
from sklearn.metrics import accuracy_score 
```

```python
acc2=accuracy_score(dep_test,dep_pred)
```

```python
acc2
```

**Using Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(ind_train,dep_train)
```

```python
dep_pred=rfc.predict(ind_test)
```
